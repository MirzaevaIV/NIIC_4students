```python
import py3Dmol
import ipywidgets as widgets
```

```python

class Smotrelka:
    def __init__(self):
        #-------------viewer-----------
        self.viewer = py3Dmol.view()
        self.viewer.setBackgroundColor('lightblue')
        
        #-------------interface-----------
        self.uploader = widgets.FileUpload(multiple=True, description='Open', button_style='success')
        self.uploader.observe(self.on_upload_change, names='_counter')
        
        self.clear_button = widgets.Button(
            description='Clear',
            disabled=False,
            button_style='danger', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Delete all objects from viewer',
            #icon='check' # (FontAwesome names without the `fa-` prefix)
        )
        self.clear_button.on_click(self.on_clear_button_clicked)
        
        self.fname_w = widgets.Label(value=' ')
        
        self.main = widgets.VBox([widgets.HBox([self.uploader, self.clear_button]),self.fname_w])
    
    def on_upload_change(self, change):
        if self.uploader._counter:
            for name, file_info in self.uploader.value.items():
                if name.split(".")[-1] == 'xyz':
                    mol_str = file_info['content'].decode("utf-8")
                    self.viewer.addModel(mol_str, "xyz")
                    self.viewer.setStyle({"stick":{"radius":0.15},"sphere": {"scale":0.25}})
                elif name.split(".")[-1] == 'cube':
                    voldata = file_info['content'].decode("utf-8")
                    self.viewer.addVolumetricData(voldata, "cube", {'isoval': -0.03, 'color': "red", 'opacity': 0.75})
                    self.viewer.addVolumetricData(voldata, "cube", {'isoval': 0.03, 'color': "blue", 'opacity': 0.75})
                self.fname_w.value += "   "+name        
            self.viewer.update()
        self.uploader.value.clear()
        self.uploader._counter=0
        
    def on_clear_button_clicked(self, b):
        self.fname_w.value = ' '
        self.viewer.removeAllModels()
        self.viewer.removeAllSurfaces()
        self.viewer.removeAllShapes()
        self.viewer.update()
        
    def show(self):
        display(self.main)
        self.viewer.show()
```

```python
Smotrelka().show()
```
